```python
def display_columns_with_few_unique_values(df, threshold=30):
    categorical_features=[]
    print("Columns with fewer than", threshold, "unique values:\n")
    found = False
    for col in df.columns:
        unique_values = df[col].unique()
        num_unique = len(unique_values)
        if num_unique < threshold:
            found = True
            categorical_features.append(col)
            print(f"Column: {col} (Unique Count: {num_unique})")
            print("Unique Values:", unique_values)
            print("-" * 50)
    if not found:
        print(f"No columns found with fewer than {threshold} unique values.")
        categorical_features=['Null']
    print('\nThese are possible Categorical Features:')
    return categorical_features

# Example usage:
# display_columns_with_few_unique_values(df, threshold=20)
```

```python
categorical_features = display_columns_with_few_unique_values(df)
categorical_features
```

### âœ… **Applying Categorical Encoding**
Since this dataset contains categorical variables, I need to convert them into numerical form before using them in a machine learning model. Different types of encoding methods can be used depending on the type of categorical data.

---

### ðŸ”¹ **Step 1: Choosing the Right Encoding**
| Column | Cardinality (Unique Values) | Recommended Encoding |
|---------|---------------------------|----------------------|
| **Damage** (6 values) | Low | **Ordinal Encoding** |
| **CAL FIRE Unit** (27 values) | Medium | **Label Encoding** |
| **Structure Type** (19 values) | Medium | **One-Hot Encoding** |
| **Roof Construction** (11 values) | Low-Medium | **One-Hot Encoding** |
| **Vent Screen** (10 values) | Low-Medium | **One-Hot Encoding** |
| **Exterior Siding** (11 values) | Low-Medium | **One-Hot Encoding** |
| **Window Pane** (7 values) | Low | **One-Hot Encoding** |

---

### ðŸ”¹ **Step 2: Implementing Encoding Methods**
#### 1ï¸âƒ£ **Ordinal Encoding for "Damage"**
The **"Damage"** column follows an increasing severity scale, ordinal encoding is the best choice:

#### 2ï¸âƒ£ **Label Encoding for CAL FIRE Unit feature Medium Cardinality**
For categorical features with greater than **20 unique values**, **Label Encoding** is ideal:

#### 3ï¸âƒ£ **One-Hot Encoding for Low-Cardinality Categorical Features**
For categorical features with fewer than **20 unique values**, **One-Hot Encoding** is ideal:

```python
import joblib
# Define encoding mappings
ordinal_categories = [["No Damage", "Affected (1-9%)", "Minor (10-25%)", "Major (26-50%)", "Destroyed (>50%)", "Inaccessible"]]

# Initialize encoders
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)
label_encoder = LabelEncoder()
one_hot_encoders = {}

# Encode ordinal column: 'Damage'
df_final["Damage"] = ordinal_encoder.fit_transform(df_final[["Damage"]])

# Encode label column: 'CAL FIRE Unit'
df_final["CAL_FIRE_Unit"] = label_encoder.fit_transform(df_final["CAL_FIRE_Unit"])

# Encode one-hot columns
one_hot_cols = ["Structure_Type", "Roof_Construction", "Vent_Screen", "Exterior_Siding", "Window_Pane"]

for col in one_hot_cols:
    one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    encoded_values = one_hot_encoder.fit_transform(df_final[[col]])
    
    # Create new column names
    new_cols = [f"{col}_{cat}" for cat in one_hot_encoder.categories_[0]]
    df_final[new_cols] = encoded_values
    
    # Drop original column
    df_final.drop(columns=[col], inplace=True)
    
    # Save the encoder
    one_hot_encoders[col] = one_hot_encoder

# Save encoders
joblib.dump(ordinal_encoder, "ordinal_encoder.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")
joblib.dump(one_hot_encoders, "one_hot_encoders.pkl")

print("Encoders saved successfully and dataset prepared for training!")
```

## **Train Models with GridSearchCV**

Training multiple models using **GridSearchCV**, evaluating them with **confusion matrix and classification report**.

---

### **ðŸ”¹ Steps in the Code**
1. **Define models and hyperparameters for GridSearchCV**  
2. **Train each model using GridSearchCV**  
3. **Evaluate the best model using**:
   - Confusion Matrix ðŸ“Š
   - Classification Report ðŸ“

---

```python
# Load dataset (Assuming X_train, X_test, y_train, y_test are already split)
# Ensure Latitude & Longitude are scaled
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[['Latitude', 'Longitude']] = scaler.fit_transform(X_train[['Latitude', 'Longitude']])
X_test_scaled[['Latitude', 'Longitude']] = scaler.transform(X_test[['Latitude', 'Longitude']])

# Define models & hyperparameters
models = {
    "LogisticRegression": (LogisticRegression(), {"C": [0.01, 0.1, 1, 10],"max_iter":[500,1000]}),
    "RandomForest": (RandomForestClassifier(), {"n_estimators": [50, 100], "max_depth": [5, 10]}),
    "GradientBoosting": (GradientBoostingClassifier(), {"n_estimators": [50, 100], "learning_rate": [0.01, 0.1]}),
    "AdaBoost": (AdaBoostClassifier(), {"n_estimators": [50, 100]}),
    "DecisionTree": (DecisionTreeClassifier(), {"max_depth": [5, 10, None]}),
}

best_models = {}

# GridSearchCV Training
for name, (model, params) in models.items():
    print(f"\nðŸ”„ Training {name}...")
    grid_search = GridSearchCV(model, param_grid=params, cv=3, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train_scaled, y_train)
    best_models[name] = grid_search.best_estimator_
    print(f"âœ… Best {name} Model: {grid_search.best_params_}")


# Save the fitted scaler
joblib.dump(scaler, "scaler.pkl")
    
# Evaluation
for name, model in best_models.items():
    print(f"\nðŸ“Œ Evaluating {name}...")
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{name} - Confusion Matrix")
    plt.show()
    
    # Classification Report
    print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred))
```

**Observation:**
- Decision Tree and Random Forest models gave better results but I will go with decision tree model and **save it**.

## Testing Input and Output from saved Decision tree model

```python
def predict_damage(input_data):
    # Load encoders and model
    label_encoder = joblib.load("label_encoder.pkl")
    one_hot_encoders = joblib.load("one_hot_encoders.pkl")
    model = joblib.load("decision_tree_model.pkl")
    
    # Convert input to DataFrame
    df_input = pd.DataFrame([input_data])

    # Encode label column: 'CAL FIRE Unit'
    df_input["CAL_FIRE_Unit"] = label_encoder.transform(df_input["CAL_FIRE_Unit"])

    # Apply scaling before prediction
    df_input[["Latitude", "Longitude"]] = scaler.transform(df_input[["Latitude", "Longitude"]])

    # Encode one-hot columns
    for col in one_hot_encoders:
        one_hot_encoder = one_hot_encoders[col]
        encoded_values = one_hot_encoder.transform(df_input[[col]])
        new_cols = [f"{col}_{cat}" for cat in one_hot_encoder.categories_[0]]
        df_encoded = pd.DataFrame(encoded_values, columns=new_cols)

        # Drop original column and merge encoded values
        df_input = pd.concat([df_input.drop(columns=[col]), df_encoded], axis=1)

    # Ensure all expected features are present
    model_features = joblib.load("encoded_data_columns.pkl")  # Save column order during training
    for col in model_features:
        if col not in df_input:
            df_input[col] = 0  # Add missing columns with zero

    # Reorder columns to match training data
    df_input = df_input[model_features]
    print('The below are the input to the model to predict: \n')
    print(df_input.loc[0])
    print('\n')
    # Make prediction
    prediction = model.predict(df_input)

    # Decode the prediction from ordinal encoding
    ordinal_encoder = joblib.load("ordinal_encoder.pkl")
    decoded_prediction = ordinal_encoder.inverse_transform([[prediction[0]]])[0][0]

    return decoded_prediction
```

```python
new_data = {
  "CAL_FIRE_Unit": "LNU",
  "Structure_Type": "Church",
  "Roof_Construction": "Wood",
  "Vent_Screen": "Unknown",
  "Exterior_Siding": "Wood",
  "Window_Pane": "No Windows",
  "Latitude": 37.7749,
  "Longitude": -122.4194,
  "Incident_Start_Year": 2020,
  "Incident_Start_Month": 9,
  "Incident_Start_Day": 15
}

prediction_result = predict_damage(new_data)
print("Predicted Damage Category:", prediction_result)
```
